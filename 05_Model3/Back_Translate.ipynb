```python
import pandas as pd
from deep_translator import GoogleTranslator
import random
import time

def back_translation(sentence, retries=3, delay=2):
    languages = ['ja', 'es', 'en']  # Japanese, Spanish, and back to English

    for attempt in range(retries):
        try:
            # English to Japanese
            translator = GoogleTranslator(source='auto', target=languages[0])
            translated = translator.translate(sentence)

            # Japanese to Spanish
            translator = GoogleTranslator(source=languages[0], target=languages[1])
            translated = translator.translate(translated)

            # Spanish back to English
            back_translator = GoogleTranslator(source=languages[1], target=languages[2])
            back_translated = back_translator.translate(translated)

            time.sleep(1)  # Add a 1-second pause between requests
            return back_translated

        except Exception as e:
            print(f"Attempt {attempt + 1} failed: {str(e)}")
            if attempt < retries - 1:
                sleep_time = delay * (2 ** attempt) + random.uniform(0, 1)
                print(f"Retrying in {sleep_time:.2f} seconds...")
                time.sleep(sleep_time)
            else:
                print("All retries failed. Skipping this sentence.")
                return sentence  # Fallback to original sentence if translation fails

df = pd.read_csv("kaggle_dataset.csv")

translated_text = df.copy()

# Apply back-translation to all rows and print row number
for index, row in translated_text.iterrows():
    print(f"Processing row {index + 1}/{len(translated_text)}")
    translated_text.at[index, 'back_translated_text'] = back_translation(row['text'])

# Save the DataFrame with back-translated text as a new CSV file
translated_text.to_csv("back_translated_dataset.csv", index=False)

# Print the first few rows to see the results
print(translated_text[['text', 'back_translated_text']].head())
```
