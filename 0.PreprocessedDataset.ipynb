```python
#Cleaning data
#1.Remove all null values with "missing" value and create 2 new columns to extract missing values. ลบค่า NULL
FN_dataset = pd.read_csv('C:/Users/Monaliza/Downloads/WELFake_Dataset_edit1.csv')

FN_dataset['title_missing'] = FN_dataset['title'].isna() | FN_dataset['title'].eq('')
FN_dataset['text_missing'] = FN_dataset['text'].isna() | FN_dataset['text'].eq('')

FN_dataset['title'].replace('', 'missing', inplace=True)
FN_dataset['text'].replace('', 'missing', inplace=True)
FN_dataset['title'].fillna('missing', inplace=True)
FN_dataset['text'].fillna('missing', inplace=True)

#2. convert eveyrthing to lowercase เปลี่ยนเป็นตัวเล็ก
FN_dataset['title'] = FN_dataset['title'].str.lower()
FN_dataset['text'] = FN_dataset['text'].str.lower()

#3 remove simple punctuation such as "" and '
def remove_punctuation(text):
    text = re.sub(f"[{string.punctuation}]", "", text)  
    text = re.sub(r"[^\w\s]", "", text)
    return text

FN_dataset['title'] = FN_dataset['title'].apply(remove_punctuation)
FN_dataset['text'] = FN_dataset['text'].apply(remove_punctuation)

# Apply the changes
FN_dataset.to_csv('C:/Users/Monaliza/Downloads/WELFake_Dataset_edit2.csv', index=False)
```

```python
#4.Removing stopwords ลบคำซ้ำ is am are this that ... etc
#Download stop words 
nltk.download('stopwords')
print(stopwords.words('english'))
```

```python
#Removing stopwords in Title and Text columns
FN_dataset = pd.read_csv(f'C:/Users/Monaliza/Downloads/WELFake_Dataset_edit2.csv') 
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    if isinstance(text, str):
        word_tokens = word_tokenize(text)
        filtered_text = [word for word in word_tokens if word.lower() not in stop_words]
        return ' '.join(filtered_text)
    return text 

FN_dataset['text'] = FN_dataset['text'].apply(remove_stopwords)
FN_dataset['title'] = FN_dataset['title'].apply(remove_stopwords)

#Apply the changes 
FN_dataset.to_csv(f'C:/Users/Monaliza/Downloads/WELFake_Dataset_edit3.csv', index=False)
```

```python
#Verify the change
FN_dataset.head()
```

```python
#5.Tokenization ทำให้ตัวอักษรเป็นคำ
#Download the NLTK tokenizer models 
nltk.download('punkt')
FN_dataset = pd.read_csv(f'C:/Users/Monaliza/Downloads/WELFake_Dataset_edit3.csv')

FN_dataset['title'] = FN_dataset['title'].fillna('').astype(str)
FN_dataset['text'] = FN_dataset['text'].fillna('').astype(str)

def tokenize_text(text):
    return [word for word in word_tokenize(text) if word.isalnum()]

#Apply tokenization to 'title' and 'text' columns
FN_dataset['title_tokens'] = FN_dataset['title'].apply(tokenize_text)
FN_dataset['text_tokens'] = FN_dataset['text'].apply(tokenize_text)

#Apply the changes 
FN_dataset.to_csv(f'C:/Users/Monaliza/Downloads/WELFake_Dataset_edit4.csv', index=False)
```

```python
#6 Lemmatization increase consistency and vocabulary เช่น based, basing, bases เป็น base ลด overfitting
FN_dataset = pd.read_csv(f'C:/Users/Monaliza/Downloads/WELFake_Dataset_edit4.csv')

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def lemmatize_text(text):
    if isinstance(text, str):
        word_tokens = word_tokenize(text)
        lemmatized_tokens = [lemmatizer.lemmatize(word.lower()) for word in word_tokens if word.lower() not in stop_words]
        return ' '.join(lemmatized_tokens)
    return text

FN_dataset['title'] = FN_dataset['title'].apply(lemmatize_text)
FN_dataset['text'] = FN_dataset['text'].apply(lemmatize_text)

#Apply the changes
FN_dataset.to_csv('C:/Users/Monaliza/Downloads/WELFake_Dataset_Finalized.csv', index=False)
```
